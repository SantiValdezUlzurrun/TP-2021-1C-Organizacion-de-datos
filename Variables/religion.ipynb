```python
plt.figure(dpi = 100)

df_sin_cristianismo = df[df['religion'] != 'cristianismo']
ax = sns.countplot(x="religion", hue="tiene_alto_valor_adquisitivo", data=df_sin_cristianismo)
plt.show()
```

```python

df_matrimonio = df[df['estado_marital'] == 'matrimonio']

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=[6.4 * 2, 4.8], dpi=100)

df_mat_alto_valor = df_matrimonio[df_matrimonio['tiene_alto_valor_adquisitivo'] == 1]
df_mat_bajo_valor = df_matrimonio[df_matrimonio['tiene_alto_valor_adquisitivo'] != 1]

df_mat_alto_valor.religion.value_counts().plot(kind='bar', ax=axes[0])
axes[0].set_title("religion - alto valor")
axes[0].set_ylabel("religion")

df_mat_bajo_valor.religion.value_counts().plot(kind='bar', ax=axes[1])
axes[1].set_title("religion - no alto valor")
axes[1].set_ylabel("religion")

plt.show()
```
